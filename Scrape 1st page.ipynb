```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name')
names
```

```python
raw_cust_name = []
for i in range(2,len(names)):
    if i not in cust_name:
        raw_cust_name.append(names[i].get_text())
raw_cust_name
```

```python
used = set()
cust_name=list(dict.fromkeys(raw_cust_name))
cust_name
```

```python
title = soup.find_all('a',class_='review-title')
title
```

```python
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate = []
for i in range(2,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
review = soup.find_all("span",class_="a-size-base review-text review-text-content")
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
lenth=(len(review_content),len(review_title),len(rate),len(review_content),len(cust_name))
lenth
```

```python
df = pd.DataFrame()
```

```python
df['Customer Name']=cust_name
df['Review title']=review_title
df['Ratings']=rate
df['Reviews']=review_content
```
